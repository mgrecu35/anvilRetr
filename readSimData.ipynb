```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
fname='simObsCrossTrack_2dn.nc'
with nc.Dataset(fname) as f:
    iwc_SSRG=f.variables['iwc_SSRG'][:]
    iwcCS=f.variables['iwcCS'][:]
```

```python
print(iwc_SSRG.mean(),iwcCS.mean())
import matplotlib
print(np.corrcoef(iwc_SSRG.flatten(),iwcCS.flatten()))
ax=plt.subplot(111)
h2d=plt.hist2d(iwc_SSRG.flatten(),iwcCS.flatten(),bins=np.arange(100)*0.01,
               norm=matplotlib.colors.LogNorm(),cmap='jet')
ax.set_aspect('equal')
```
